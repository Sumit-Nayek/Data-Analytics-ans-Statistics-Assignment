```python
## 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dataset
data = {
    "Employee": ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10'],
    "Department": ['HR', 'IT', 'Marketing', 'Finance', 'IT', 'HR', 'Marketing', 'Finance', 'IT', 'HR'],
    "Experience": [2, 5, 1, 8, 4, 6, 3, 10, 7, 5],
    "Salary": [35000, 55000, 32000, 70000, 48000, 52000, 41000, 82000, 60000, 54000]
}
df = pd.DataFrame(data)

# Boxplot for salary distribution across departments
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Department', y='Salary')
plt.title('Salary Distribution Across Departments')
plt.show()

# Histogram for salary frequency
plt.figure(figsize=(8, 6))
plt.hist(df['Salary'], bins=8, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Salaries')
plt.xlabel('Salary ($)')
plt.ylabel('Frequency')
plt.show()

# Salary variation by department
variation = df.groupby('Department')['Salary'].std()
print("Salary Variation by Department:")
print(variation)
```

```python
## 2
# Dataset
data = {
    "Student": ['S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10'],
    "Age": [10, 12, 11, 13, 10, 12, 11, 13, 10, 12],
    "Height": [140, 150, 145, 160, 138, 152, 147, 165, 143, 155]
}
df = pd.DataFrame(data)

# Scatter plot for Age vs Height
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Height'], color='blue', alpha=0.7)
plt.title('Age vs Height')
plt.xlabel('Age (Years)')
plt.ylabel('Height (cm)')
plt.show()

# Density plot for Age
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Age'], fill=True, color='green')
plt.title('Density Plot of Age')
plt.xlabel('Age (Years)')
plt.show()

# Boxplot for Height
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Height'], color='orange')
plt.title('Height Distribution')
plt.ylabel('Height (cm)')
plt.show()
```

```python
## 3
# Dataset
data = {
    "Product": ['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10'],
    "Category": ['Electronics', 'Furniture', 'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics', 'Clothing', 'Electronics'],
    "Region": ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    "Sales Volume": [5000, 3000, 7000, 4500, 6000, 8000, 4000, 3500, 9000, 3000],
    "Profit (%)": [10, 15, 20, 12, 18, 22, 9, 16, 25, 11]
}
df = pd.DataFrame(data)

# Bar plot for total sales volume by category
plt.figure(figsize=(8, 6))
df.groupby('Category')['Sales Volume'].sum().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Sales Volume by Category')
plt.ylabel('Sales Volume')
plt.xlabel('Category')
plt.show()

# Multiple boxplots for profit percentage by category
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Category', y='Profit (%)', hue='Region')
plt.title('Profit Percentage by Category and Region')
plt.show()

# Average profit by category
avg_profit = df.groupby('Category')['Profit (%)'].mean()
print("Average Profit by Category:")
print(avg_profit)
```

```python
## 4
# Dataset
data = {
    "Customer": ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10'],
    "Age": [22, 35, 27, 45, 34, 29, 50, 42, 33, 38],
    "Monthly Expenditure": [250, 450, 300, 550, 400, 320, 600, 520, 390, 480]
}
df = pd.DataFrame(data)

# Bubble plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Monthly Expenditure'], s=df['Age']*10, alpha=0.7, color='purple')
plt.title('Bubble Plot: Age vs Monthly Expenditure')
plt.xlabel('Age (Years)')
plt.ylabel('Monthly Expenditure ($)')
plt.show()

# Density plot for Age
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Age'], fill=True, color='red')
plt.title('Density Plot of Age')
plt.xlabel('Age (Years)')
plt.show()

# Trend analysis
plt.scatter(df['Age'], df['Monthly Expenditure'], alpha=0.7)
plt.title('Age vs Expenditure Trend')
plt.xlabel('Age (Years)')
plt.ylabel('Monthly Expenditure ($)')
plt.show()
```

```python
## 5
# Dataset
data = {
    "Month": ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
    "Avg Temperature": [5, 6, 10, 15, 20, 25, 28, 27, 23, 15, 10, 7],
    "Precipitation": [80, 70, 85, 60, 55, 40, 30, 45, 65, 75, 85, 90]
}
df = pd.DataFrame(data)

# Line plot for temperature changes
plt.figure(figsize=(8, 6))
plt.plot(df['Month'], df['Avg Temperature'], marker='o', linestyle='-', color='blue')
plt.title('Average Temperature Over Months')
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.show()

# Scatter plot for temperature vs precipitation
plt.figure(figsize=(8, 6))
plt.scatter(df['Avg Temperature'], df['Precipitation'], color='green')
plt.title('Temperature vs Precipitation')
plt.xlabel('Avg Temperature (C)')
plt.ylabel('Precipitation (mm)')
plt.show()
```

```python

# Find months with highest temperatures and least precipitation
highest_temp_months = df.loc[df['Avg Temperature'] == df['Avg Temperature'].max(), 'Month'].tolist()
lowest_precip_months = df.loc[df['Precipitation'] == df['Precipitation'].min(), 'Month'].tolist()

print(f"Months with highest average temperature: {highest_temp_months}")
print(f"Months with lowest precipitation: {lowest_precip_months}")

# Insights:
#  - The months with highest temperatures generally correspond to the months with the least precipitation. This suggests a relationship where higher temperatures lead to less precipitation.
#  - This could indicate a drier climate during warmer months.
#  - Further analysis could include calculating correlation coefficients or performing statistical tests to confirm these relationships.

# Enhanced Visualization
plt.figure(figsize=(10, 6))

# Plot Temperature
plt.plot(df['Month'], df['Avg Temperature'], marker='o', linestyle='-', color='blue', label='Avg Temperature')

# Plot Precipitation (on a secondary y-axis for better comparison)
plt.twinx()  # Create a second y-axis
plt.plot(df['Month'], df['Precipitation'], marker='s', linestyle='--', color='green', label='Precipitation')

plt.title('Average Temperature and Precipitation Over Months')
plt.xlabel('Month')
plt.ylabel('Temperature (C)', color='blue')  # Label the primary y-axis
plt.ylabel('Precipitation (mm)', color='green', rotation=-90, va="bottom") # Label the secondary y-axis
plt.legend(loc='upper left') # Position the legend


plt.show()
```
